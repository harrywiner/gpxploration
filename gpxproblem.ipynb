```python
import folium
import gpxpy
```

```python
# Read GPX file
with open('./data/St-Rumbolds-Way.gpx', 'r') as gpx_file:
    gpx = gpxpy.parse(gpx_file)
```

```python
gpx.routes[1]
```

```python
first_point = gpx.routes[0].points[0]
m = folium.Map(location=[first_point.latitude, first_point.longitude], zoom_start=14)

# Plot track
for route in gpx.routes:
    color = None
    if route.name == "StRumwoldsWay.gpx_ALT Rumbold route":
        color = "red"
    else:
        color = "blue"
    points = [(point.latitude, point.longitude) for point in route.points]
    folium.PolyLine(points, color=color, weight=2.5, opacity=0.5).add_to(m)

# Display map
# m.save('GPX_track_map.html')
m  # i
```
